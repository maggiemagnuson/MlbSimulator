```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.baseball-reference.com/leagues/MLB/2019.shtml"
page = requests.get(url, verify=False)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find('table', {'id': 'teams_standard_batting'})
```

```python
rows = []
for tr in table.find_all('tr'):
    row = []
    try:
        row.append(tr.find_all('a')[0].text)
    except:
        try:
            row.append(tr.find_all('em')[0].text)
        except: 
            row.append("TOTAL")

    for td in tr.find_all('td'):
        row.append(td.text.replace('\n', ''))
        if len(row) == 28:
            rows.append(row)
        
```

```python
columns=["Tm","NumPlayers","BatAge","RunsPerGame","G","PA","AB","R","H","2B","3B","HR","RBI","SB","CS","BB","SO","BA","OBP","SLG","OPS","OPSAdjusted","TB","GDP","HBP","SH","SF","IBB","LOB"]
batting_df = pd.DataFrame(rows, columns=columns)
```

```python
batting_df.tail()
```

```python
url = "https://www.baseball-reference.com/leagues/MLB/2019-pitches-pitching.shtml"
page = requests.get(url, verify=False)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
pitches_pitching_table = soup.find('table', {'id': 'teams_pitches_pitching'})
rows = []
for tr in pitches_pitching_table.find_all('tr'):
    row = []
    try:
        row.append(tr.find_all('a')[0].text)
    except:
        try:
            row.append(tr.find_all('em')[0].text)
        except: 
            row.append("TOTAL")

    for td in tr.find_all('td'):
        row.append(td.text.replace('\n', ''))
        if len(row) == 31:
            rows.append(row)
```

```python
columns = ["Tm","RunAllowedPerGame","PlateAppearances","Pit","PitPerPlateAppearance","Strikes","StrikePct","StrikeLookingPct","StrikeSwingingPct","StrikeFoulPct","BallsInPlayPct","StrikeSwungPct","IntentionalBallPct","PitchesSwungAtPct","ContactPct","1stPitchStrikePct","30CountSeenPct","30CountSeen","30StikeCount","02CountSeenPct","02CountSeen","02StrikeCount","02HitsAllowed","StrikeoutLooking","StrikeoutSwinging","StrikeoutLookingPct","3PitchStrikeout","4PitchWalk","PAu","Pitu","Stru"]
pitches_pitching_df = pd.DataFrame(rows, columns=columns)
```

```python
pitches_pitching_df.tail()
```

```python
#Pitching Calculations
tmp_innings_pitched = pitching_df.iloc[31]["IP"]
tmp_innings_pitched
innings_pitched, partial_innings_pitched = tmp_innings_pitched.split(".")
print(innings_pitched)
ip = int(innings_pitched) + (int(partial_innings_pitched) * .33)
batters_faced = ((ip*2.82)+int(pitching_df.iloc[31]["H"])+int(pitching_df.iloc[31]["BB"]))/int(batting_df.iloc[31]["G"])
print(batters_faced)
aveNumOfPitchesAve = batters_faced * float(pitches_pitching_df.iloc[31]['PitPerPlateAppearance'])
```

```python
onBaseAve = (int(pitching_df.iloc[31]["H"])+int(pitching_df.iloc[31]["BB"]))/(int(batting_df.iloc[31]["AB"])+int(batting_df.iloc[31]["BB"]))
```
